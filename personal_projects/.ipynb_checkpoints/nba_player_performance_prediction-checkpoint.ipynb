```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import RobustScaler
from scipy.stats.mstats import winsorize
from scipy.stats import percentileofscore
# import statsmodels.api as sm
# from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns

sns.set()
```

## Load raw data

```python
raw_data = pd.read_csv("../NBA/nba.csv")
raw_data.describe(include="all")
```

## Check for missing values

```python
if raw_data.isna().any().any():
    data_no_mv = raw_data.dropna()
else:
    data_no_mv = raw_data
```

## Preprocessing

```python
data_no_mv["Season_type"] = data_no_mv["Season_type"].str.replace("Regular%20Season", "Regular_Season")
data_no_mv.head()
```

## Log transformation to get normality of key metrics and deal with outliers

```python
data_transformed = data_no_mv.copy()

# List of columns to transform
columns_to_transform = ["PTS", "AST", "TOV", "STL", "BLK", "REB", "PF", "FGA", "FG3A", "FTA"]

for col in columns_to_transform:
    data_transformed[f"LOG_{col}"] = np.log(data_no_mv[col] + 1)

# Winsorize the log transformed data
for col in columns_to_transform:
    data_transformed[f"LOG_{col}"] = winsorize(data_transformed[f"LOG_{col}"], limits=[0.05, 0.01])

# Robust scale the data
robust_scaler = RobustScaler()
data_transformed[[f"LOG_{col}" for col in columns_to_transform]] = robust_scaler.fit_transform(data_transformed[[f"LOG_{col}" for col in columns_to_transform]])

sns.histplot(data_transformed["LOG_FG3A"], stat="density")
```

## View OLS assumptions on scatter plot

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_no_mv["AST"], data_transformed["PTS"])
ax1.set_title("Points and Assists")

ax2.scatter(data_no_mv["TOV"], data_transformed["PTS"])
ax2.set_title("Points and Turnovers")

ax3.scatter(data_no_mv["FG3A"], data_transformed["PTS"])
ax3.set_title("Points and 3 Pt Field Goals Attenpted")

plt.show()
```
