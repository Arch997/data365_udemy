## Import the relevant libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm 
from sklearn.linear_model import LinearRegression
import seaborn as sns 

sns.set()
```

## Load the data

```python
data = pd.read_csv("real_estate_price_size_year.csv")
data.head()
```

```python
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price']
x1 = data[["size", "year"]]

# Create the scikit model
model = LinearRegression()
model.fit(x1, y)
```

### Regression

```python
# Add constant
x = sm.add_constant(x1)


results = sm.OLS(y, x).fit()
results.summary()
```

```python
# Add the intercepts and the independent variables to get separate yhats

yhat_size = 5772000 + 227.7009 * x1["size"]
yhat_year = 5772000 + 2916.7853 * x1["year"]

print(yhat_size, yhat_year)
```

```python
# Plot the relationship between Y and X1
plt.scatter(data["size"], y)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Price", fontsize=15)

# Show the regression line
fig = plt.plot(data['size'], yhat_size, lw=4, c="orange", label="regression line")

# Show the plot
plt.show()

plt.scatter(x1["year"], y)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Price", fontsize=15)

# Show the regression line
fig = plt.plot(x1['year'], yhat_year, lw=4, c="yellow", label="regression line")
plt.show()



# plot.xlabel("
```
